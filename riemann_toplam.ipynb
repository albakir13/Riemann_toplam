```python
def f(x):
    return x**3

import numpy as np

def riemann_toplam(f, a, b, N, yontem):
    dx = (b-a) / N
    
    x = np.linspace(a,b,N+1)
    
    if yontem == 'alt_riemann':
        return np.sum(f(x[:-1])*dx)
    elif yontem == 'ust_riemann':
        return np.sum(f(x[1:])*dx)
```

```python
riemann_toplam(f,0,1,100,'alt_riemann')
```

```python
riemann_toplam(f,0,1,100,'ust_riemann')
```
